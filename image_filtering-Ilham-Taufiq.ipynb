```python
#Menyiapkan modul yang diperlukan
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('sample_data/OIP.jpg')
#menampilkan gambar
cv2_imshow(img)

#membuat filter: matriks berukuran 5x5
kernel = np.ones((5,5),np.float32)/25
print(kernel)
```

```python
#menampilkan gambar setelah diubah dari bgr ke rgb
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img1)
```

```python
#filtering blur pada gambar original
img_filter = cv2.filter2D(img, -1, kernel)
cv2_imshow(img_filter)
```

```python
#plot gambar original
plt.subplot(121),plt.imshow(img1),plt.title('Original')
plt.xticks([]), plt.yticks([])

#plot gambar hasil filter
plt.subplot(122),plt.imshow(img_filter),
plt.title('Averaging')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#Mengecek nomor pallete gambar yang di filter
b, g, r = cv2.split(img_filter)

print('Dimensi Gambar')
print(img.shape)
```

```python
#menampilkan gambar dengan grayscale
img2 = cv2.imread('sample_data/OIP.jpg',0)

#high-pass filtering laplacian
laplacian = cv2.Laplacian(img2,cv2.CV_64F)
#CV_64F menunjukkan nilai bit yang dihasilkan dengan tipe data float

sobelx = cv2.Sobel(img2,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img2,cv2.CV_64F,0,1,ksize=5)

#memperbesar ukuran hasil plotting
plt.rcParams["figure.figsize"] = (20,20)

#Menampilkan filtering
plt.subplot(2,2,1),plt.imshow(img2,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show
```

```python
#menampilkan gambar dengan grayscale
img3 = cv2.imread('sample_data/OIP.jpg',0)

#menggunakan fungsi Canny Edges dengan parameter (img3, nilai_min, nilai_max)
edges = cv2.Canny(img3,100,200)

plt.subplot(121),plt.imshow(img3,cmap = 'gray')
plt.title('Original'),plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge'),plt.xticks([]),plt.yticks([])

plt.show()
```

```python
#Hitungan treshold.
#menampilkan gambar dengan grayscale
img4 = cv2.imread('sample_data/OIP.jpg',0)

ret,trshold1 = cv2.threshold(img4,127,255,cv2.THRESH_BINARY)
ret,trshold2 = cv2.threshold(img4,127,255,cv2.THRESH_BINARY_INV)
ret,trshold3 = cv2.threshold(img4,127,255,cv2.THRESH_TRUNC)
ret,trshold4 = cv2.threshold(img4,127,255,cv2.THRESH_TOZERO)
ret,trshold5 = cv2.threshold(img4,127,255,cv2.THRESH_TOZERO_INV)

#Hasil
titl = ['Original', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
imgs = [img4, trshold1, trshold2, trshold3, trshold4, trshold5]

#Menampilkan gambar
for i in range(6):
  plt.subplot(3,2,i+1),plt.imshow(imgs[i], 'gray')
  plt.title(titl[i])
  plt.xticks([]),plt.yticks([])

plt.show()
```
